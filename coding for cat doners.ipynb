The Cat Connection is an all-volunteer nonprofit organization whose primary mission is to rescue abandoned, abused, unwanted, stray, neglected, and injured cats, and to facilitate their adoption into responsible permanent homes. You volunteered to help them develop a small Python-based system that allows them to store, search, and manage their donor information.
The program reads in existing donor information from a data file (donors.csv) and another file which is the donor transactions(transactions.txt). The first row is the column headings and the rest of rows are the records in each file.
The program reads the information from the file to memory when it runs and writes the modified information back from memory to the file when the user exits the program. All changes to the information remain in memory while the program runs.




required library import

```python
import pandas as pd
from csv import writer
import os
import re
```

```python
def display_donars(donar):
    x=donar.shape
    col=donar.columns
    print('                                              Donor Names                              ')
    print('------------------------------------------------------------------------------------------------------------')
    print(f'{col[0]:4}        {col[1]:10}        {col[2]:10}        {col[3]:25}        {col[4]:10}         {col[5]:5}')
    print('------------------------------------------------------------------------------------------------------------')
    
    for i in range(x[0]):
        row=list(donar.iloc[i,:])
        print(f'{row[0]:4}        {row[1]:10}        {row[2]:10}        {row[3]:25}        {row[4]:10}         {row[5]:5}')
   
    print('============================================================================================================')
def display_donations(amount):
    x=amount.shape
    col=amount.columns
    print('                                 Donors and Amounts Donated')
    print('-------------------------------------------------------------------------------')
    print(f'{col[1]:10}        {col[2]:10}        {col[4]:10}        {col[5]:5}       {col[6]:10}')
    print('-------------------------------------------------------------------------------')
    
    for i in range(x[0]):
        row=list(amount.iloc[i,:])
        last='$'+str(row[6])
        print(f'{row[1]:10}        {row[2]:10}        {row[4]:10}        {row[5]:5}       {last:10}')
   
    print('===============================================================================')
    total=amount['Amount'].sum()
    tot='$'+str(total)
    print(f"Total Donations                                                   {tot:}")
    
def find_largest_donation(amount):
    max_amt=amount['Amount'].max()
    max_donar=amount[amount['Amount']==max_amt]
    x=max_donar.shape
    for i in range(x[0]):
        
        name=list(max_donar.iloc[i,:])
        
        print("The largest donation is ${} from {} {}".format(max_amt,name[1],name[2]))
    
def isValidEmail(email):
    regex = r'\b[A-Za-z0-9._%+-]+@[A-Za-z0-9.-]+\.[A-Z|a-z]{2,}\b'
    if(re.fullmatch(regex, email)):
        return True
    else:
        return False
def add_donar(donar):
    data=[]
    lst_id=list(donar['ID'])
    data=[lst_id[-1]+10]
    fn=list(donar['FirstName'])
    ln=list(donar['LastName'])
    
    
    name=[]
    for i in range(len(fn)):
        name.append(str(fn[i])+' '+str(ln[i]))
    en=0
    while en==0:
        fname=input("Enter the first name: ")
        lastn=input("Enter the last name: ")
        gname=fname+' '+lastn
        if gname not in name:
            en=1
        if en==0:
            print("User name already exists!")
    em=0
    while em==0:
        email=input("Enter the email address: ")
        if isValidEmail(email):
            em=1
        if em==0:
            print("Invalid email-ID!")
        
    data.append(fname)
    data.append(lastn)
    data.append(email)
    data.append(input("Enter the city: "))
    data.append(input("Enter the state: "))
    try:
        file = 'donors.csv'
        os.remove(file)
       
        donar.loc[len(donar.index),:] = data
        donar.to_csv('donors.csv', index=False)
    except:
        print("Error in file update:")
def remove_donar(donar):
    fname=input("Enter the first name: ")
    lastn=input("Enter the last name: ")
    
    op=1
    try:
        donar.drop(donar.loc[(donar['FirstName'] ==fname) & (donar['LastName'] == lastn)].index, inplace=True)
        file = 'donors.csv'
        os.remove(file)     
        donar.to_csv('donors.csv', index=False)
        op=0
    except:
        print("Error in donar removing!")

    if op==1:
        print('Donar name not Exists')

def add_a_donation(donar,trans):
    fname=input("Enter the first name: ")
    lastn=input("Enter the last name: ")
    
    fn=list(donar['FirstName'])
    ln=list(donar['LastName'])
    dt=[]
    name=[]
    for i in range(len(fn)):
        name.append(str(fn[i])+' '+str(ln[i]))
   # print(name)
    fnmm=str(fname)+' '+(lastn)
   # print(fnmm)
    try:
        if fnmm in name:
            idx=name.index(fnmm)
            ids=list(donar.iloc[idx,:])
            donar_id=ids[0]
        amt=input("Enter the amount of the donation:  ")
        dat=input("Enter the date of the donation:   ")
        dt=[donar_id,amt,dat]
        try:
          file = 'transactions.csv'
          os.remove(file)
        
          trans.loc[len(trans.index),:] = dt
          
        
              
          trans.to_csv('transactions.csv', index=False)
        except:
            print("Error in donation update:") 
            
    except:
        print("Not a valid donar name!")
    
      
```

```python
adef main():        
    while True:

        print("""****************  Donor Menu  *****************
        1 - Display Donors            4 - Add a Donor
        2 - Display Donor Amounts     5 - Add a Donation
        3 - Display Largest Amount    6 - Remove a Donor
        Q - Quit""")
        
        donar=pd.read_csv('donors.csv')
        trans=pd.read_csv('transactions.csv')
        amount = pd.merge(donar, trans, on='ID')
        
        option=input("Enter choice: ")
        
        if option=='1':
            display_donars(donar)
            
        elif option=='2':
            display_donations(amount)   
            
        elif option=='3':
            find_largest_donation(amount)
            
        elif option=='4':
            print("Add a Donar")
            add_donar(donar)
            
        elif option=='5':
            print("Add a Donation")
            add_a_donation(donar,trans)
            
        elif option=='6':
            
            print("Remove a Donar")
            remove_donar(donar)
        elif option=='Q' or 'q':
            break
        else:
            pass
        
    
   
if __name__=="__main__":
    main()
```
